PART 1


Importing Libraries

```python
import sklearn
from sklearn import datasets
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

Loading breast cancer dataset and dimension reductionality

```python
b_cancer = datasets.load_breast_cancer()
data_bcancer = b_cancer.data

pca_2feature = PCA(2)
bcdf = pca_2feature.fit_transform(data_bcancer)
```

Performing KMEANS Clustering (10 Clusters)

```python
kmeans =  KMeans(n_clusters = 10)
label = kmeans.fit_predict(bcdf)
```

```python
for i in range(10):
    plt.scatter(bcdf[label==i,0],bcdf[label==i,1], label = i )
plt.legend()
plt.show()
```

PART 2

Loading Digits Dataset and Dimension reductionality

```python
digi = datasets.load_digits()
dig_data = digi.data

pca_2feature = PCA(2)
dg = pca_2feature.fit_transform(dig_data)
```

Performing DBSCAN


```python
dbscan =  DBSCAN(min_samples=10, eps=1.5)
label = dbscan.fit_predict(dg)
```

```python
plot_uniquecolors = ['midnightblue', 'indigo', 'magenta','deeppink', 'pink', 'khaki', 'cyan', 'slategray', 'plum', 'lightcoral', 'darkcyan', 'powderblue','blue','lightgrey','maroon', 'red', 'sandybrown', 'linen', 'olive', 'yellow', 'lawngreen','green', 'teal', 'steelblue' ]
for i in range(23):
    plt.scatter(dg[label==i,0],dg[label==i,1], color = plot_uniquecolors[i], label = i)
plt.legend( loc='right', bbox_to_anchor=(1.6, 0.5), ncol=3)
plt.show()
```

PART 3


Loading Iris Dataset and Dimension Reductionality

```python
iris = datasets.load_iris()
iris_data=iris.data

pca_2feature = PCA(2)
dfi = pca_2feature.fit_transform(iris_data)
```

Agglomerative Clustering on Iris Dataset ( 5 Clusters)

```python
agg =  AgglomerativeClustering(n_clusters = 5)
label = agg.fit_predict(dfi)
```

```python
for i in range(5):
    plt.scatter(dfi[label==i,0],dfi[label==i,1], label = i+1)
plt.legend()
plt.show()
```
